```python
import pandas as pd
from datetime import datetime
import calendar
from pandas import to_datetime
%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\Jojo3\Downloads\Project 2 Copy 1\delhi_weather.csv',skipinitialspace=True)
```

# Cleaning

```python
df = df.rename(columns={'WindDirDegrees<br />':'WindDirDegrees'})
```

```python
df['WindDirDegrees'] = df['WindDirDegrees'].str.rstrip('<br />')
```

```python
df['WindDirDegrees'] = df['WindDirDegrees'].astype('float64') 
```

```python
df['Events'] = df['Events'].fillna('')
```

```python
df['Date'] = to_datetime(df['Date'])
```

To determine the best two weeks for a summer break, you would likely want to consider columns that are related to weather and climate conditions. Based on the given DataFrame columns, some important columns that you may want to consider are:
